```python
import os, cv2
import numpy as np
import pandas as pd
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

import albumentations as A
from albumentations.pytorch import ToTensorV2

import timm

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
```

```python
IMG_DIR1 = "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000/HAM10000_images_part_1/"
IMG_DIR2 = "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000/HAM10000_images_part_2/"
CSV_PATH = "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000/HAM10000_metadata.csv"
```

```python
df = pd.read_csv(CSV_PATH)

df['path'] = df['image_id'].apply(lambda x:
    os.path.join(IMG_DIR1, x+'.jpg') if os.path.exists(os.path.join(IMG_DIR1, x+'.jpg'))
    else os.path.join(IMG_DIR2, x+'.jpg')
)

le = LabelEncoder()
df['label'] = le.fit_transform(df['dx'])

NUM_CLASSES = df['label'].nunique()
```

```python
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(df['label']),
    y=df['label']
)

class_weights = torch.tensor(class_weights, dtype=torch.float).cuda()
```

```python
train_aug = A.Compose([
    A.Resize(224, 224),

    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),

    A.Affine(
        scale=(0.9, 1.1),
        translate_percent=(0.05, 0.05),
        rotate=(-30, 30),
        shear=(-10, 10),
        p=0.7
    ),

    A.OneOf([
        A.MotionBlur(blur_limit=5),
        A.MedianBlur(blur_limit=5),
        A.GaussianBlur(blur_limit=5),
    ], p=0.3),

    A.RandomBrightnessContrast(p=0.5),

    A.HueSaturationValue(
        hue_shift_limit=10,
        sat_shift_limit=15,
        val_shift_limit=10,
        p=0.5
    ),

    A.CoarseDropout(
        num_holes_range=(1, 8),
        hole_height_range=(0.05, 0.2),
        hole_width_range=(0.05, 0.2),
        fill=0,
        p=0.5
    ),

    A.Normalize(),
    ToTensorV2()
])

valid_aug = A.Compose([
    A.Resize(224, 224),
    A.Normalize(),
    ToTensorV2()
])
```

```python
class SkinDataset(Dataset):
    def __init__(self, df, aug):
        self.df = df.reset_index(drop=True)
        self.aug = aug

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        img = cv2.imread(row['path'])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        img = self.aug(image=img)['image']
        label = torch.tensor(row['label'], dtype=torch.long)

        return img, label
```

```python
class Model(nn.Module):
    def __init__(self, backbone):
        super().__init__()
        self.model = timm.create_model(backbone, pretrained=True, num_classes=NUM_CLASSES)

    def forward(self, x):
        return self.model(x)
```

```python
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
```

```python
def mixup(x, y, alpha=0.4):
    lam = np.random.beta(alpha, alpha)
    index = torch.randperm(x.size(0)).cuda()

    mixed_x = lam * x + (1 - lam) * x[index]
    y_a, y_b = y, y[index]

    return mixed_x, y_a, y_b, lam
```

```python
def tta(model, img):
    imgs = [
        img,
        torch.flip(img, [2]),
        torch.flip(img, [3])
    ]

    preds = []
    for im in imgs:
        preds.append(torch.softmax(model(im), dim=1))

    return torch.mean(torch.stack(preds), dim=0)
```

```python
from torch.amp import autocast, GradScaler
from tqdm import tqdm

def train_fold(df, train_idx, val_idx, backbone, fold, img_size=224, batch_size=6, epochs=5):

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Update augmentations for Kaggle
    train_aug_kaggle = A.Compose([
        A.Resize(img_size, img_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.Normalize(),
        ToTensorV2()
    ])

    valid_aug_kaggle = A.Compose([
        A.Resize(img_size, img_size),
        A.Normalize(),
        ToTensorV2()
    ])

    train_loader = DataLoader(
        SkinDataset(df.iloc[train_idx], train_aug_kaggle),
        batch_size=batch_size,
        shuffle=True,
        num_workers=4,
        pin_memory=True
    )

    val_loader = DataLoader(
        SkinDataset(df.iloc[val_idx], valid_aug_kaggle),
        batch_size=batch_size,
        shuffle=False,
        num_workers=4,
        pin_memory=True
    )

    model = Model(backbone).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=5, T_mult=2)
    scaler = GradScaler()

    best_acc = 0

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        # Freeze head for first 2 epochs
        if epoch < 2:
            for name, p in model.named_parameters():
                p.requires_grad = "head" in name or "fc" in name
        else:
            for p in model.parameters():
                p.requires_grad = True

        for batch_idx, (img, label) in enumerate(tqdm(train_loader, desc=f"Fold {fold} Epoch {epoch}")):
            img, label = img.to(device), label.to(device)
            img, y_a, y_b, lam = mixup(img, label)

            optimizer.zero_grad()
            with autocast(device_type="cuda"):
                pred = model(img)
                loss = lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

            scaler.scale(loss).backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            if batch_idx % 10 == 0:
                print(f"Batch {batch_idx}/{len(train_loader)} Loss {running_loss/(batch_idx+1):.4f}")

        scheduler.step()

        # VALIDATION (no TTA to speed up)
        model.eval()
        preds, labels_list = [], []
        with torch.no_grad():
            for img, label in val_loader:
                img = img.to(device)
                out = torch.softmax(model(img), dim=1)  # simple forward pass
                p = torch.argmax(out, dim=1)
                preds.extend(p.cpu().numpy())
                labels_list.extend(label.numpy())

        acc = accuracy_score(labels_list, preds)
        print(f"{backbone} Fold {fold} Epoch {epoch} Accuracy: {acc:.4f}")

        # Save best model
        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), f"{backbone}_fold{fold}.pth")
            print(f"Saved best model for fold {fold} at epoch {epoch}")
```

```python
from torch.amp import autocast, GradScaler
from tqdm import tqdm

def train_fold(df, train_idx, val_idx, backbone, fold, img_size=224, batch_size=6, epochs=5, tta=False):

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Augmentations
    train_aug_kaggle = A.Compose([
        A.Resize(img_size, img_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.Normalize(),
        ToTensorV2()
    ])

    valid_aug_kaggle = A.Compose([
        A.Resize(img_size, img_size),
        A.Normalize(),
        ToTensorV2()
    ])

    train_loader = DataLoader(
        SkinDataset(df.iloc[train_idx], train_aug_kaggle),
        batch_size=batch_size,
        shuffle=True,
        num_workers=0,        # avoid freeze
        pin_memory=True
    )

    val_loader = DataLoader(
        SkinDataset(df.iloc[val_idx], valid_aug_kaggle),
        batch_size=batch_size,
        shuffle=False,
        num_workers=0,        # avoid freeze
        pin_memory=True
    )

    model = Model(backbone).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=5, T_mult=2)
    scaler = GradScaler()

    best_acc = 0

    for epoch in range(epochs):
        model.train()

        # Freeze backbone for first 2 epochs
        if epoch < 2:
            for name, p in model.named_parameters():
                p.requires_grad = "head" in name or "fc" in name
        else:
            for p in model.parameters():
                p.requires_grad = True

        running_loss = 0.0
        for batch_idx, (img, label) in enumerate(tqdm(train_loader, desc=f"Fold {fold} Epoch {epoch}")):
            img, label = img.to(device), label.to(device)
            img, y_a, y_b, lam = mixup(img, label)

            optimizer.zero_grad()
            with autocast(device_type="cuda"):
                pred = model(img)
                loss = lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

            scaler.scale(loss).backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            if batch_idx % 50 == 0:
                print(f"Batch {batch_idx}/{len(train_loader)} Loss {running_loss/(batch_idx+1):.4f}")

        scheduler.step()

        # VALIDATION
        model.eval()
        preds, labels_list = [], []
        with torch.no_grad():
            for img, label in val_loader:
                img = img.to(device)
                if tta:
                    out = tta(model, img)
                else:
                    out = torch.softmax(model(img), dim=1)
                p = torch.argmax(out, dim=1)
                preds.extend(p.cpu().numpy())
                labels_list.extend(label.numpy())

        acc = accuracy_score(labels_list, preds)
        print(f"{backbone} Fold {fold} Epoch {epoch} Accuracy: {acc:.4f}")

        # Save best model
        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), f"{backbone}_fold{fold}.pth")
            print(f"Saved best model for fold {fold} at epoch {epoch}")
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

backbones = [
    'tf_efficientnet_b6.ns_jft_in1k',
    'convnextv2_large.fcmae_ft_in22k_in1k',
    'eva_large_patch14_448'
]

oof_preds = np.zeros((len(df), NUM_CLASSES))
oof_labels = np.zeros(len(df))

for backbone in backbones:
    for fold, (train_idx, val_idx) in enumerate(skf.split(df, df['label'])):

        train_fold(df, train_idx, val_idx, backbone, fold)

        model = Model(backbone).cuda()
        model.load_state_dict(torch.load(f"{backbone}_fold{fold}.pth"))
        model.eval()

        val_loader = DataLoader(
            SkinDataset(df.iloc[val_idx], valid_aug),
            batch_size=16,
            shuffle=False
        )

        idx_ptr = 0

        with torch.no_grad():
            for img, label in val_loader:
                img = img.cuda()

                out = tta(model, img).cpu().numpy()

                bs = len(img)
                indices = val_idx[idx_ptr:idx_ptr+bs]

                oof_preds[indices] += out
                oof_labels[indices] = label.numpy()

                idx_ptr += bs
```

```python
# ===== Cell 1: EfficientNet B6 =====
backbone = 'tf_efficientnet_b6.ns_jft_in1k'

oof_preds = np.zeros((len(df), NUM_CLASSES))
oof_labels = np.zeros(len(df))

skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)  # 2 folds for Kaggle

for fold, (train_idx, val_idx) in enumerate(skf.split(df, df['label'])):
    print(f"Training {backbone} Fold {fold}")

    train_fold(df, train_idx, val_idx, backbone, fold)  # Your training function

    # Load best checkpoint
    model = Model(backbone).cuda()
    model.load_state_dict(torch.load(f"{backbone}_fold{fold}.pth"))
    model.eval()

    val_loader = DataLoader(
        SkinDataset(df.iloc[val_idx], valid_aug),
        batch_size=6,
        shuffle=False,
        num_workers=4,
        pin_memory=True
    )

    idx_ptr = 0
    with torch.no_grad():
        for img, label in val_loader:
            img = img.cuda()
            out = tta(model, img).cpu().numpy()  # Can skip TTA to speed up

            bs = len(img)
            indices = val_idx[idx_ptr:idx_ptr+bs]

            oof_preds[indices] += out
            oof_labels[indices] = label.numpy()
            idx_ptr += bs

# Save EfficientNet predictions
np.save("oof_preds_efficientnet.npy", oof_preds)
np.save("oof_labels.npy", oof_labels)
```

```python
# ===== Cell: ConvNeXtV2 Small/Base Training =====
from torch.amp import autocast, GradScaler
from tqdm import tqdm

# Choose a smaller variant for Kaggle Free GPU
backbone = 'convnextv2_base.fcmae_ft_in22k_in1k'  # or 'convnextv2_small'

oof_preds = np.zeros((len(df), NUM_CLASSES))
oof_labels = np.zeros(len(df))

skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)  # 2 folds for speed

def train_fold_fast(df, train_idx, val_idx, backbone, fold, img_size=224, batch_size=6, epochs=5, tta=False):

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Lightweight augmentations
    train_aug_kaggle = A.Compose([
        A.Resize(img_size, img_size),
        A.HorizontalFlip(p=0.5),
        A.Normalize(),
        ToTensorV2()
    ])

    valid_aug_kaggle = A.Compose([
        A.Resize(img_size, img_size),
        A.Normalize(),
        ToTensorV2()
    ])

    train_loader = DataLoader(
        SkinDataset(df.iloc[train_idx], train_aug_kaggle),
        batch_size=batch_size,
        shuffle=True,
        num_workers=0,  # avoid Kaggle CPU freeze
        pin_memory=True
    )

    val_loader = DataLoader(
        SkinDataset(df.iloc[val_idx], valid_aug_kaggle),
        batch_size=batch_size,
        shuffle=False,
        num_workers=0,
        pin_memory=True
    )

    model = Model(backbone).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=5, T_mult=2)
    scaler = GradScaler()

    best_acc = 0

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        # Freeze backbone for first 2 epochs
        if epoch < 2:
            for name, p in model.named_parameters():
                p.requires_grad = "head" in name or "fc" in name
        else:
            for p in model.parameters():
                p.requires_grad = True

        for batch_idx, (img, label) in enumerate(tqdm(train_loader, desc=f"Fold {fold} Epoch {epoch}")):
            img, label = img.to(device), label.to(device)
            img, y_a, y_b, lam = mixup(img, label)

            optimizer.zero_grad()
            with autocast(device_type="cuda"):
                pred = model(img)
                loss = lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

            scaler.scale(loss).backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            if batch_idx % 50 == 0:
                print(f"Batch {batch_idx}/{len(train_loader)} Loss {running_loss/(batch_idx+1):.4f}")

        scheduler.step()

        # VALIDATION
        model.eval()
        preds, labels_list = [], []
        with torch.no_grad():
            for img, label in val_loader:
                img = img.to(device)
                if tta:
                    out = tta(model, img)
                else:
                    out = torch.softmax(model(img), dim=1)
                p = torch.argmax(out, dim=1)
                preds.extend(p.cpu().numpy())
                labels_list.extend(label.numpy())

        acc = accuracy_score(labels_list, preds)
        print(f"{backbone} Fold {fold} Epoch {epoch} Accuracy: {acc:.4f}")

        # Save best model
        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), f"{backbone}_fold{fold}.pth")
            print(f"Saved best model for fold {fold} at epoch {epoch}")

# ===== Train 2 folds =====
for fold, (train_idx, val_idx) in enumerate(skf.split(df, df['label'])):
    print(f"Training {backbone} Fold {fold}")
    train_fold_fast(df, train_idx, val_idx, backbone, fold, img_size=224, batch_size=6, epochs=5, tta=False)

    # Load best model and save OOF predictions
    model = Model(backbone).cuda()
    model.load_state_dict(torch.load(f"{backbone}_fold{fold}.pth"))
    model.eval()

    val_loader = DataLoader(
        SkinDataset(df.iloc[val_idx], valid_aug),
        batch_size=16,
        shuffle=False,
        num_workers=0,
        pin_memory=True
    )

    idx_ptr = 0
    with torch.no_grad():
        for img, label in val_loader:
            img = img.cuda()
            out = torch.softmax(model(img), dim=1).cpu().numpy()
            bs = len(img)
            indices = val_idx[idx_ptr:idx_ptr+bs]
            oof_preds[indices] += out
            oof_labels[indices] = label.numpy()
            idx_ptr += bs

# Save OOF predictions
np.save("oof_preds_convnext_fast.npy", oof_preds)
np.save("oof_labels.npy", oof_labels)
```
